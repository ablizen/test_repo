```python
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

### Создаем csv файлы для обучения и теста

```python
data = datasets.make_regression(n_samples=1000, n_features=5, n_targets=1, bias = 0.5, coef=True)
```

```python
data
```

```python
X = data[0]
y = data[1]
```

```python
data[0].shape
```

```python
colnames = ["col"+str(i) for i in range(5)]
colnames.append("target")
```

```python
colnames
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
train_data = pd.DataFrame(data= np.c_[X_train, y_train],
                     columns= colnames)

test_data = pd.DataFrame(data= np.c_[X_test, y_test],
                     columns= colnames)
```

```python
train_data.to_csv("train_data.csv", header=True, index=False)
```

```python
test_data.to_csv("test_data.csv", header=True, index=False)
```
